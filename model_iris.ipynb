```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree


```

```python
iris = load_iris()
x= iris.data
y = iris.target
#print(iris.feature_names)

#separation 80/20 des donnes pour entrainement/test
train_X, test_X, train_Y, test_Y = train_test_split(x, y, test_size=0.2,stratify=y)

#instanciation d'un modelle de type arbre decisionnel
model = DecisionTreeClassifier(max_depth=3)

#entrainement du model
model.fit(train_X, train_Y)

#test du model
y_pred = model.predict(test_X)

#affichage du score de justesse
print(accuracy_score(test_Y, y_pred)*100)

fig = plt.figure(10)
plot_tree(model,filled=True,max_depth=3,feature_names=iris.feature_names,class_names=iris.target_names)
plt.show()
```
